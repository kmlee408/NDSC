```python
import os
import sys
import time
import multiprocessing as mp
_sys_add_path = os.path.abspath('./YT_api')
sys.path.insert(1,_sys_add_path)

import pandas as pd

import YouTube
```

```python
fpath = './datasets/USvideos.csv'
_dt = pd.read_csv(fpath)

_vids = _dt['video_id']
l_vids = len(_vids)


yt = YouTube.YouTube('AIzaSyDdPiRW0lXGtY_s188gr6tIZ8WY_ZqAglM')
```

```python
def runner_get_duration(vid_list):
    _duration = [ yt.get_duration(vid) for vid in vid_list ]
    return _duration
```

```python
_st = time.time()


result = []
try:
    MP = 40
    pool = mp.Pool(processes=MP)

    _div = int(l_vids / MP) + 1 #2000
    #_parts = int(l_vids / _div) #_div개씩, parts = 20개

    for idx in range(MP):# 20번 반복
        _s = idx*_div
        _e = _s + _div

        lst = _vids[_s:_e]
        result.append(pool.apply_async(runner_get_duration, args=(lst,)))
    
except Exception as e:
    print(e)
finally:
    pool.close()
    pool.join()
    
durations = [] #will be list of lists
for rst in result:
    durations.append(rst.get()) #append a list
    
    
_et = time.time()
print('time elapsed : '+str(_et-_st))
```
